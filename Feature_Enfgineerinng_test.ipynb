```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Display maximum columns
pd.set_option('display.max_columns',None)
df=pd.read_csv('test.csv')
df.head()
```

## Missing Values


```python
#Lets capure missing values in categorical features first
categorical_nan=[features for features in df.columns if df[features].isnull().sum()>1 and df[features].dtype=='O']
for features in categorical_nan:
    print(f'{features} has {np.round(df[features].isnull().mean(),4)} % null values')
```

```python
# Replace missing values with new labels
def replace_cat_na(df,categorical_nan):
    data=df.copy()
    data[categorical_nan]=data[categorical_nan].fillna('Missing')
    return data

df=replace_cat_na(df,categorical_nan)

df[categorical_nan].isnull().sum()
```

```python
df.head()
```

```python
# lets check missing values for numerical features
numerical_nan=[features for features in df.columns if features not in categorical_nan and df[features].isnull().sum()>1]

for features in numerical_nan:
    print(f'{features} has {np.round(df[features].isnull().mean(),4)} % null values')
```

```python
# Replacing null values in numerical features
for features in numerical_nan:
    #we will replace with median because columns have outliers
    median_value=df[features].median()

    #Capture null values in new column
    df[features+' nan']=np.where(df[features].isnull(),1,0)
    df[features]=df[features].fillna(median_value)
    
print(df[numerical_nan].isnull().sum())
```

```python
df.head(10)
```

```python
# Temporal Cariables(Date time variables)
for features in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[features]=df['YrSold']-df[features]
 
```

### Numerical Features
Since numerical features are skewed we will perform log normal distribution


```python
df.head()
```

```python
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for features in num_features:
    df[features]=np.log(df[features])  #Performing log normal distribution on numerical features
```

```python
df.head()
```

### Handling Rare Categorical values
we will remove categorical values that are present in less than 1% of observations


```python
categorical_features=[features for features in df.columns if df[features].dtype=='O']
print(categorical_features)
```

```python
for features in categorical_features:
    temp=df.groupby(features).count()/len(df)
    temp_df=temp[temp>0.01].index
    df[features]=np.where(df[features].isin(temp_df),df[features],'Rar value')
```

```python
df.head()
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature]).mean().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head()
```

### Feature Scaling
We will perform feature scaling on all features excluding id and Saleprice with MinMaxScaler


```python
feature_scale=[features for features in df.columns if features not in ['Id']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
# Concatenating Id and SalePrice with feature scale dataframe
data=pd.concat([df[['Id']].reset_index(drop=True),
                            pd.DataFrame(scaler.transform(df[feature_scale]),columns=feature_scale)],axis=1)
```
